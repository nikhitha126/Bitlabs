```python
import matplotlib.pyplot as plt
import pandas as pd

```

```python
months=['jan','feb','march','april','may','june']
savings=[10000,20000,30000,40000,50000,60000]
plt.plot(months,savings,linestyle="dashed",color="green",marker='*')
plt.title("Customized Line Plot")
plt.xlabel('months')
plt.ylabel('savings')
```

```python
plt.bar(months,savings,color=['green','blue','pink','red','yellow','violet'])
plt.title("Customized Bar Plot")
plt.xlabel('months')
plt.ylabel('savings')
```

```python
import matplotlib.pyplot as plt

age = [22, 34, 23, 24, 25,29,36, 25]
plt.hist(age, bins=5)

plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['Python', 'Java', 'C++', 'JavaScript', 'Ruby']
sizes = [40, 30, 15, 10, 5]
plt.pie(sizes,labels=labels, autopct='%1.1f%%')
plt.title("Programming Languages")
plt.show()
```
